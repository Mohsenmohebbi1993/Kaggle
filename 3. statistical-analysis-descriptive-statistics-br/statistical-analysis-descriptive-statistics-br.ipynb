#### Step 1 - Specify the limits for each class

<b>A</b> ‚ñ∫ above 20 SM

<b>C</b> ‚ñ∫ from 4 to 10 SM

<b>D</b> ‚ñ∫ from 2 to 4 SM

onde <b>SM</b> is the minimum wage at the time. In our case <b>R$ 788,00</b> (2015):

<b>B</b> ‚ñ∫ from 7.880 to 15.760

<b>C</b> ‚ñ∫ from 3.152 to 7.880

<b>D</b> ‚ñ∫ from 1.576 to 3.152

<b>E</b> ‚ñ∫ Up until 1.576


```python
dados.Renda.unique()
```

```python
dados["Renda"][0]
```

```python
dados.head(2)
renda_type = dados.copy()
def categorize_renda(Randa):
    if Randa > 15760:
        return "A"
    elif Randa > 7880:
        return "B"
    elif Randa > 3152:
        return "B"
    elif Randa > 1576:
        return "C"
    elif Randa > 0:
        return "D"
    else:
        return "E"


renda_type["categorize"] = renda_type["Renda"].apply(categorize_renda)
renda_categorize = pd.DataFrame(renda_type.groupby(["categorize"])["categorize"].count())
renda_categorize["Percent"] = (renda_categorize / np.sum(renda_categorize))*100
renda_categorize
```

```python
# Coluna com a frequ√™ncia das classes por renda
freq_renda = table(cut(x=dados$Renda,
    breaks<-c(0, 1576, 3152, 7880, 15760, 200000),
    labels<-c('E', 'D', 'C', 'B', 'A'),
    include.lowest=T))

# Coluna com o percentual das classes por renda
perct_renda = prop.table(table(cut(x=dados$Renda,
    breaks<-c(0, 1576, 3152, 7880, 15760, 200000),
    labels<-c('E', 'D', 'C', 'B', 'A'),
    include.lowest=T))) * 100
freq_perct_renda <- data.frame(cbind("Frequency"=freq_renda, "Percent"=perct_renda))


# Visualizando a matriz com a frequ√™ncia e o percentual de cada classe
freq_perct_renda[order(row.names(freq_perct_renda)),]
```

### Frequency distribution for quantitative variables (fixed amplitude classes)

---

> ## Sturges Rule
> ## $K = 1 + \frac{10}{3}log_{10}n$

---

> The Sturges rule is a criterion used to determine the number of classes or intervals needed to graph a set of statistical data. This rule was enunciated in 1926 by the German mathematician Herbert Sturges.

> <a href="https://maestrovirtuale.com/regra-de-sturges-explicacao-aplicacoes-e-exemplos/">MORE INFORMATION...</a>

## E. Central tendency measures

<a id="central"></a>

><b> In statistics, a central trend (or, normally, a measure of central trend) is a central value or typical value for a probability distribution. The most common measures of central tendency are the arithmetic mean, the median and the mode. Central trends can be calculated both for a finite number of values and for a theoretical distribution, like the normal distribution. Occasionally authors use a central tendency (or centrality), meaning "the tendency of quantitative data to group around a central value." Such meaning can be expected from the usual definition of the words trend and centrality in the dictionary. Authors can judge whether data has a strong or weak central tendency based on statistical dispersion, measured by standard deviation or something similar.</b>

### Median

> Median is the value that separates the major and minor half of a sample, population or probability distribution. In simpler terms, median can be the middle value of a data set. In the data set {1, 3, 3, 6, 7, 8, 9}, for example, the median is 6.

To obtain the median of a data set, we must proceed as follows:
1. Sort the dataset;
2. Identify the number of observations (records) in the data set ($n$);
3. Identify the middle element:

> Quando $n$ for √≠mpar, a posi√ß√£o do elemento mediano ser√° obtida da seguinte forma:


## $$Elemento_{Md} = \frac{n+1}2$$

> When $n$ is odd, the position of the median element will be obtained as follows:


## $$element_{Md} = \frac{n}2$$

4. Get the median:

> When $ n $ is odd:


## $$Md = X_{Element_{Md}}$$

> When $ n $ is even:


## $$Md = \frac{X_{Element_{Md}} + X_{Element_{Md}+1}}2$$

> Using the <b>R</b> language, we only need to insert the command median, whose parameter that the median function receives are all records of the attribute from which we want to obtain the median. Below is the use of the <b>median()</b> function, to obtain the medians of each quantitative attribute of the database under analysis.
***

```python
df_medianas = data.frame(list(Medianas=c(median(dados$Idade), median(dados$Anos.de.Estudo), median(dados$Renda), median(dados$Altura))))
row.names(df_medianas) = c("Idade", 'Anos.de.Estudo', 'Renda', 'Altura')
df_medianas
```

### Mode

><b> Fashion is one of the measures of height of a data set, as well as the mean and the median. It can be defined in sample and population fashion. In relation to the first one, the sampling mode of a data set deals with the value that occurs most frequently or the most common value in a data set.</b>

```python
# Fun√ß√£o para obt√™n√ß√£o da moda
Moda = function(x){
    
    freq <- table(x)
    return(names(freq)[freq == max(freq)])
}
```

## G. Separating Measures

> There are a number of position measures similar in their design to the median, although they are not measures of central tendency. As is known, the median divides the distribution into two equal parts in terms of the number of elements in each part. Quartiles, on the other hand, allow the distribution to be divided into four equal parts in terms of the number of elements in each; the deciles in ten parts and the centiles in one hundred equal parts. Basically the separating measures are three, <b>quartiles</b>, <b>deciles</b> and <b>percentiles</b>.
<a id="separatrizes"></a>

### Quartiles - Renda

> In descriptive statistics, a quartile is any of the three values that divide the ordered set of data into four equal parts, and so each part represents <b>1/4</b> of the sample or population.

```python
quartis<-c()
for(i in 1:3){
    
    quartis<-c(quartis, i/4)
}
df_quartis<-data.frame(quantile(dados$Renda, quartis))
colnames(df_quartis) <- c("Quartiles")
df_quartis
```

### deciles - Renda

```python
decis<-c()
for(i in 1:9){
    
    decis<-c(decis, i/10)
}
df_decis<-data.frame(quantile(dados$Renda, decis))
colnames(df_decis) <- c("Deciles")
df_decis
```

### Percentis - Renda

> In descriptive statistics, percentiles are measures that divide the sample (in ascending order of data) into 100 parts, each with an approximately equal percentage of data. The k-th percentile Pk is the x (xk) value that corresponds to the cumulative frequency of N. k / 100, where N is the sample size.

## H. Dispersion Measures

> Although the position measures provide a very important summary of the data, they may not be sufficient to characterize different sets, especially when the observations of a given distribution present very dispersed data.

<a id="dispers√£o"></a>

### Mean absolute deviation</font>
***


> ## $$DM = \frac 1n\sum_{i=1}^{n}|X_i-\bar{X}|$$

```python
df_ab = data.frame("Mean absolute deviation"=c(MeanAD(dados$Renda), MeanAD(dados$Idade), MeanAD(dados$Anos.de.Estudo), MeanAD(dados$Altura)))
row.names(df_ab) <- c("Renda", "Idade", "Anos de Estudo", "Altura")
df_ab
```

### Variance

> The variance is constructed from the differences between each observation and the data mean, that is, the deviation around the mean. In calculating the variance, the deviations around the mean are squared.

### Population variance

## $$\sigma^2 = \frac 1n\sum_{i=1}^{n}(X_i-\mu)^2$$

### Sample variance

## $$S^2 = \frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$$

```python
df_var = data.frame("variance"=c(var(dados$Renda), var(dados$Idade), var(dados$Anos.de.Estudo), var(dados$Altura)))
row.names(df_var) <- c("Renda", "Idade", "Anos de Estudo", "Altura")
df_var
```

### Standard deviation

><b> One of the restrictions of variance is the fact that it provides square measures of the original units - the variance of length measures, for example, is in units of area. Therefore, the fact that the units are different makes it difficult to compare dispersion with the variables that define it. One way to eliminate this difficulty is to consider its square root.</b>

### Population standard deviation

## $$\sigma = \sqrt{\frac 1n\sum_{i=1}^{n}(X_i-\mu)^2} \Longrightarrow \sigma = \sqrt{\sigma^2}$$

### Sample standard deviation

## $$S = \sqrt{\frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2} \Longrightarrow S = \sqrt{S^2}$$

```python
df_std = data.frame("Standard deviation"=c(sqrt(var(dados$Renda)), sqrt(var(dados$Idade)), sqrt(var(dados$Anos.de.Estudo)), sqrt(var(dados$Altura))))
row.names(df_std) <- c("Renda", "Idade", "Anos de Estudo", "Altura")
df_std
```
