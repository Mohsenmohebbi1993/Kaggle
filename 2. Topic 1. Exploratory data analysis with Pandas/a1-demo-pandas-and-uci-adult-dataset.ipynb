**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
# You code here
# more and equal 50K salary
ages_1 = data.loc[data["salary"].eq(">50K"), "age"]
# less 50K salary
ages_2 = data.loc[data["salary"].eq("<=50K"), "age"]
print(f"more 50K : mean age is {ages_1.mean()} and std is {ages_1.std()}")
print(f"less and equal 50K : mean age is {ages_2.mean()} and std is {ages_2.std()}")

```

```python
ages1 = data.loc[data['salary'] == '>50K', 'age']
ages2 = data.loc[data['salary'] == '<=50K', 'age']
print("The average age of the rich: {0} +- {1} years, poor - {2} +- {3} years.".format(
    round(ages1.mean()), round(ages1.std(), 1),
    round(ages2.mean()), round(ages2.std(), 1)))
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
# You code here
# What is the maximum number of hours a person works per week (*hours-per-week* feature)
max_hore_work_per_week = data["hours-per-week"].max()
print("the maximum number of hours a person works per week is {0} hore".format(max_hore_work_per_week))

#How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?
number_workaholics = data.loc[data["hours-per-week"].eq(max_hore_work_per_week)].shape[0]
print("Total number of such hard workers {0}".format(number_workaholics))

rich_share = float(data[(data["hours-per-week"].eq(max_hore_work_per_week))
                 & (data["salary"].eq(">50K"))].shape[0]) / number_workaholics
print("Percentage of rich among them {0}%".format(int(100 * rich_share)))
```
